### Dataset 1: "Fundamentals"

*Reading in the "fundamentals" dataset that is a part of the zip file with historical data from 2012-2016 of the NY Stock Exchange, and converting it dataset into a pandas dataframe.*

```python
df = pd.read_csv('C:\\Users\\Elizabeth\\Downloads\\fundamentals.csv')
df
```

*Dropping the columns that are not needed for analysis to answer the questions above.*

```python
df = df.drop(axis=1, columns=['Accounts Payable', 'Total Current Assets', 'Accounts Receivable', "Add'l income/expense items", 'After Tax ROE', 'Capital Expenditures', 'Capital Surplus', 'Cash Ratio', 'Cash and Cash Equivalents', 'Changes in Inventories', 'Common Stocks', 'Cost of Revenue', 'Current Ratio', 'Deferred Asset Charges', 'Deferred Liability Charges', 'Depreciation', 'Earnings Before Interest and Tax', 'Earnings Before Tax', 'Effect of Exchange Rate', 'Equity Earnings/Loss Unconsolidated Subsidiary', 'Fixed Assets', 'Goodwill', 'Gross Margin', 'Income Tax', 'Intangible Assets', 'Interest Expense', 'Inventory', 'Investments', 'Liabilities', 'Long-Term Debt', 'Long-Term Investments', 'Minority Interest', 'Misc. Stocks', 'Net Borrowings', 'Net Cash Flow', 'Net Cash Flow-Operating', 'Net Cash Flows-Financing', 'Net Cash Flows-Investing', 'Net Income', 'Net Income Adjustments', 'Net Income Applicable to Common Shareholders', 'Net Income-Cont. Operations', 'Net Receivables', 'Non-Recurring Items', 'Operating Income', 'Operating Margin', 'Other Assets', 'Other Current Assets', 'Other Current Liabilities', 'Other Equity', 'Other Financing Activities', 'Other Investing Activities', 'Other Liabilities', 'Other Operating Activities', 'Other Operating Items', 'Pre-Tax Margin', 'Pre-Tax ROE', 'Profit Margin', 'Quick Ratio', 'Research and Development', 'Retained Earnings', 'Sale and Purchase of Stock', 'Sales, General and Admin.', 'Short-Term Debt / Current Portion of Long-Term Debt', 'Short-Term Investments', 'Total Assets', 'Total Current Liabilities', 'Total Equity', 'Total Liabilities & Equity', 'Treasury Stock', 'For Year', 'Earnings Per Share', 'Estimated Shares Outstanding'])
df
```

*Dropping the 'Unnamed' column that was at the beginning of the cleaned dataframe.*

```python
df.drop(df.columns[df.columns.str.contains('Unnamed',case = False)],axis = 1, inplace = True)
df
```

*Checking if there are any NA values that need to be addressed in the cleaned dataset by creating another df without NA values.*

```python
clean_df = df.dropna()
print(clean_df)
```

**There are no NA values in the updated dataset, since the number of rows stayed the same in the clean_df as the df.**

*Saving cleaned dataframe to a new csv file.*

```python
df.to_csv('Fundamentals_Cleaned.csv')
```
